```python
class MinStack:

    def __init__(self):
        self.stack = []
        self.min_stack = []

    def push(self, val: int) -> None:
        self.stack.append(val)
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self) -> None:
        if self.stack[-1] == self.min_stack[-1]:
            self.min_stack.pop()
        self.stack.pop()

    def top(self) -> int:
        return self.stack[-1]

    def getMin(self) -> int:
        return self.min_stack[-1]
```

```python
class BrowserHistory:

    def __init__(self, homepage: str):
        self.homepage = homepage
        self.stack_back = []
        self.stack_forward = []

    def visit(self, url: str) -> None:
        self.stack_back.append(self.homepage)
        self.homepage = url
        self.stack_forward = []

    def back(self, steps: int) -> str:
        while steps > 0 and self.stack_back:
            self.stack_forward.append(self.homepage)
            self.homepage = self.stack_back.pop()
            steps-=1
        return self.homepage

    def forward(self, steps: int) -> str:
        while steps > 0 and self.stack_forward:
            self.stack_back.append(self.homepage)
            self.homepage = self.stack_forward.pop()
            steps-=1
        return self.homepage
```
